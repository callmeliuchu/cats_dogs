```python
import torch
```

```python
m = 4
k = 3
n = 5
```

```python
U = [torch.rand(k,requires_grad=True) for _ in range(m)]
V = [torch.rand(k,requires_grad=True) for _ in range(n)]
```

```python
U
```

```python
V
```

```python
target = [[0.1,None,0.7,None,None],
          [None,0.5,None,None,None],
          [0.3,0.5,None,0.9,0.1],
          [None,0.3,None,0.4,0.2]]
```

```python
for _ in range(10000):
    loss = 0
    for i in range(len(target)):
        for j in range(len(target[0])):
            if target[i][j] is not None:
                loss += (torch.dot(U[i],V[j])-target[i][j])**2
    print(loss)
    loss.backward()
    for u in U:
        u.data -= 0.001*u.grad.data
        u.grad.data.zero_()
    for v in V:
        v.data -= 0.001*v.grad.data
        v.grad.data.zero_()
```

```python
mat = []
for i in range(m):
    arr = []
    for j in range(n):
        arr.append(torch.dot(U[i],V[j]).item())
    mat.append(arr)
mat = np.array(mat)
```
