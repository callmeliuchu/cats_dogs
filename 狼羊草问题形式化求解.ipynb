```python
#狼，猎人，男人，男孩子1，男孩子2，女人，女孩子1，女孩子2  船
# 0, 1,   2,   3,      4,     5,    6,      7     8
status = [1,1,1,1,1,1,1,1]

def pprint(status):
    names = ['狼','猎人','男人','男孩子1','男孩子2','女人','女孩子1','女孩子2','船']
    origin = []
    to = []
    for i,idx  in enumerate(status):
        if idx == 1:
            origin.append(names[i])
        else:
            to.append(names[i])
#     print(status)
    print('出发点',origin,'----->','对岸',to)

    
```

```python
def legal_status(status):
    # 猎人和狼不在一边
    if status[0] != status[1]:
        for i in [2,3,4,5,6,7]:
            if status[i] == status[0]:
                return False
    #女人和男人不在一边
    if status[2] != status[5]:
        # 女人和男人的孩子在一起
        if status[5] == status[3] or status[5] == status[4]:
            return False
        # 男人和女人的孩子在一起
        if status[2] == status[6] or status[2] == status[7]:
            return False
    return True
```

```python
def next_status(status):
    if status[-1] == 1:
        v1 = 1
        v2 = 0
    else:
        v1 = 0
        v2 = 1 
    ret = []
    for k in [1,2,5]:
        if status[k] != v1:
            continue
        for i in range(len(status)-1):
            if status[i] == v2:
                continue
            nums1 = list(status)
            if nums1[i] == v1:
                nums1[i] = v2
            nums1[k] = v2
            nums1[-1] = v2
            if nums1 not in ret and legal_status(nums1):
                ret.append(nums1)
    return ret
```

```python
n_s = next_status([1,1,1,1,1,1,1,1,1])
n_s
```

```python
for v in n_s:
    if legal_status(v):
        print(v)
```

```python
result = []
bg = (1,1,1,1,1,1,1,1,1)
visited = [bg]
def dfs(status,count=0):
    if status == (0,0,0,0,0,0,0,0,0):
        print('successful')
        result.append(visited[:])
        return
#     print(count,op,status)
#     pprint(status)
#     print('-'*50)
    n_s = next_status(status)
    for s in n_s:
        ss = tuple(s)
        if ss not in visited:
            visited.append(ss)
            dfs(ss,count+1)
            visited.pop()
#         else:
#             print('已存在',kk)
dfs(bg)
```
